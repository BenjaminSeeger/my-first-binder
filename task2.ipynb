## Importieren des MNIST-Datasets
Die folgenden Zeilen importieren das Dataset und weitere Dinge, welche für beispielsweise spätere Analysen hilreich sind.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Loading the MNIST-Dataset
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_validation, y_validation) = mnist.load_data()
```

## Daten vorbereiten
Durch das vorbereiten der Daten kann der Trainingsprozess beschleunigt werden, indem die Daten von den Graustufenwerten 0 bis 255 auf 0 bis 1 skaliert werden. **Später durch auskommentieren nochmal prüfen, ob das wirklich der Fall ist. Geg. weiter alle Pixel auf entweder weiß oder schwarz setzen.**

```python
x_train = x_train/255 
x_validation = x_validation/255
```

## Definition der Modelarchitektur
**Hier erklären warum ich was verwendet habe. Zuerst noch rausfinden was für meinen Fokus am besten funktioniert. Siehe auch dem file mit meinen Notizen in diesem repository.**

```python
# Definition of the model
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    
    # Hier verschiedenste Sachen noch gegenüberstellen und probieren #################################
    # Version 1 ######################################################################################
    #tf.keras.layers.Dense(512, activation=tf.nn.relu),
    #tf.keras.layers.Dense(512, activation=tf.nn.relu),
    #tf.keras.layers.Dense(512, activation=tf.nn.relu),
    #tf.keras.layers.Dense(512, activation=tf.nn.relu),
    # Version 2 ######################################################################################
    #tf.keras.layers.Dense(512, activation=tf.nn.relu),
    #tf.keras.layers.Dropout(0.5),
    #tf.keras.layers.Dense(512, activation=tf.nn.relu),
    #tf.keras.layers.Dropout(0.5),
    #tf.keras.layers.Dense(512, activation=tf.nn.relu),
    #tf.keras.layers.Dropout(0.5),
    #tf.keras.layers.Dense(512, activation=tf.nn.relu),
    #tf.keras.layers.Dropout(0.5),
    # Version 2 ######################################################################################    
    #tf.keras.layers.Dense(512, activation=tf.nn.relu, kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    #tf.keras.layers.Dense(512, activation=tf.nn.relu, kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    #tf.keras.layers.Dense(512, activation=tf.nn.relu, kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    #tf.keras.layers.Dense(512, activation=tf.nn.relu, kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    # Hier verschiedenste Sachen noch gegenüberstellen und probieren #################################
    
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

# Compile the model an set the training parameters
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

## Trainieren des Netzes
Siehe folgenden Code.

```python
import datetime
import os
logdir = os.path.join("logs", datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
tensorboard_callback = tf.keras.callbacks.TensorBoard(logdir, histogram_freq=1)

model.fit(x_train, 
          y_train, 
          epochs=10, 
          shuffle=True, 
          callbacks=[tensorboard_callback],
          validation_data=(x_validation, y_validation)
)
```

```python
import math

predictions = model.predict([x_validation])
predictions = np.argmax(predictions, axis=1)

numbers_to_display = 196
num_cells = math.ceil(math.sqrt(numbers_to_display))
plt.figure(figsize=(15, 15))

for plot_index in range(numbers_to_display):    
    predicted_label = predictions[plot_index]
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    color_map = 'Greens' if predicted_label == y_validation[plot_index] else 'Reds'
    plt.subplot(num_cells, num_cells, plot_index + 1)
    plt.imshow(x_validation[plot_index].reshape((28, 28)), cmap=color_map)
    plt.xlabel(predicted_label)

plt.subplots_adjust(hspace=1, wspace=0.5)
plt.show()
```
